#### 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
#### Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
#### You need to find following details:
#### A) Rank
#### B) Name
#### C) Artist
#### D) Upload date
#### E) Views

```python
# import useful libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
# open chromedriver using path
driver = webdriver.Chrome("I:\Software//chromedriver.exe")
```

```python
# open url
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
driver.maximize_window() # maximize driver window
time.sleep(4)  # sleep code for 4 second
```

```python
# create empty lisst to store the data
rank = []
video_name = []
artist = []
date = []
views = []
```

```python
# scrap rank of videos
rank_ = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]")
for i in rank_:
    rank.append(i.text)
    
# scrape video name
video_name_ = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]")
for i in video_name_:
    video_name.append(i.text)
    
# scrape artist name
artist_ = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]")
for i in artist_:
    artist.append(i.text)

# scrape video upload date
date_ = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]")
for i in date_:
    date.append(i.text)

# scrape views
views_ = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]")
for i in views_:
    views.append(i.text)
    
print(rank)
print(video_name)
print(artist)
print(date)
print(views)
```

```python
name = []
for x in video_name:
    name.append(x.split("[")[0])
print(name)
```

```python
# close driver
driver.close()
```

```python
# create dataframe
df = pd.DataFrame({})
df['Rank'] = rank
df['Video_name'] = name
df['Artist'] = artist
df['Upload_date'] = date
df['Views_in_billions'] = views
df
```

####  5. Scrape the details of trending repositories on Github.com.
#### Url = https://github.com/
#### You have to find the following details:
#### A) Repository title
#### B) Repository description
#### C) Contributors count
#### D) Language used
#### Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# open chromedriver using path
driver = webdriver.Chrome("I:\Software//chromedriver.exe")
```

```python
# open url
driver.get("https://github.com/")

driver.maximize_window() # maximize driver window
driver.implicitly_wait(10) # wait for open all elements of the page
```

```python
# click on explore and trending
driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary").click()
driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary").click()
driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a").click()
```

```python
# create empty list to store data
title=[]
des=[]
language=[]
count=[]
URL=[]

# scrape repositories title
title_ = driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/h1/a")
for i in title_:
    title.append(i.text)

# scrape language, repositories description and contributeros count
urls = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a") # collecting urls of all the repositories
for i in urls:
    URL.append(i.get_attribute('href')) #  getting the url of all repositories
for url in URL:
    driver.get(url)
    
    try:
        language_ = driver.find_element_by_xpath("//div[@class='BorderGrid-row']/div/ul/li[1]/a/span[1]")
        language.append(language_.text)
        
    except NoSuchElementException as e:
        language.append("No data") # appending the "No data" if data is not is there

    
    try:
        des_ = driver.find_element_by_xpath("//*[@id='repo-content-pjax-container']/div/div[2]/div[2]/div/div[1]/div/p")
        des.append(des_.text)              
        
    except NoSuchElementException as e:
        des.append("No data") # appending the "No data" if data is not is there
    
    try:
        count_ = driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span") 
        count.append(count_.text)      
                                
    except NoSuchElementException as e:
        count.append("No data") # appending the "No data" if data is not is there

print(title)
print(des)
print(language)
print(count)
```

```python
# close driver
driver.close()
```

```python
# create dataframe
df = pd.DataFrame({})
df['Repository_title'] = title
df['Repository_description'] = des
df['Language_used'] = language
df['Contributors_count'] = count
df
```

```python
# open url
driver.get("https://archive.ics.uci.edu/")
driver.maximize_window() # maximize driver window
time.sleep(4)  # sleep code for 4 second
```

```python
# click on view all datasets
driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b").click()
```

```python
# create empty list to store data
dataset_name=[]
data_type=[]
task=[]
attribute_type=[]
no_of_instances=[]
no_of_attribute=[]
year=[]

# scrape dataset name
dataset_name_ = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
for i in dataset_name_:
    dataset_name.append(i.text)

# scrape data type
data_type_ = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p") 
for i in data_type_:
    if i.text is None:
        data_type.append("--")
    else:
        data_type.append(i.text)

# scrape task
task_ = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p")
for i in task_:
    if i.text is None:
        task.append("--")
    else:
        task.append(i.text)
        
# scrape type of attribue
attribute_type_ = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
for i in attribute_type_:
    if i.text is None:
        attribute_type.append("--")
    else:
        attribute_type.append(i.text)
        
# scrape number of instances
no_of_instances_ = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
for i in no_of_instances_:
    if i.text is None:
        no_of_instances.append("--")
    else:
        no_of_instances.append(i.text)
        
# scrape number of attributes
no_of_attribute_ = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p")
for i in no_of_attribute_:
    if i.text is None:
        no_of_attribute.append("--")
    else:
        no_of_attribute.append(i.text)
        
# scrape year
year_ = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
for i in year_:
    if i.text is None:
        year.append("--")
    else:
        year.append(i.text)
        
print(dataset_name)        
print(data_type)
print(task)
print(attribute_type)
print(no_of_instances)
print(no_of_attribute)
print(year)
```

```python
# delete extra column name     
data_type.pop(0)
task.pop(0)
attribute_type.pop(0)
no_of_instances.pop(0)
no_of_attribute.pop(0)
year.pop(0)
```

```python
# close driver
driver.close()
```
