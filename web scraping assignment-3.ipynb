# web scraping assignment-3

#### 1. Write a python program which searches all the product under a particular product from www.amazon.in. The product name to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.
#### 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Import useful libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
# open chromedriver using path
driver = webdriver.Chrome("I:\Software//chromedriver.exe")
```

```python
driver.get("https://www.amazon.in/") # open url

driver.maximize_window() # maximize driver window
time.sleep(4)  # sleep code for 4 second

# enter watch in search field
driver.find_element_by_id("twotabsearchtextbox").send_keys("watch")

# click on search icon
driver.find_element_by_id("nav-search-submit-button").click()

driver.implicitly_wait(10) # wait for open all elements of the page
```

```python
# click on page-1
driver.find_element_by_xpath("//ul[@class='a-pagination']/li[2]").click()
time.sleep(4)  # sleep code for 4 second
```

```python
# extract current url.
driver.current_url
```

```python
# extract data of first page
URL_1=[]
Rating_1=[]
Return_1=[]
Avia_1=[]
Details_1=[]
No_ratings_1=[]
price_1=[]
Delivery_1=[]
Brand_1=[]
Product_name_1=[]


urls = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")  # collecting urls of all the watches
for i in urls:
    URL_1.append(i.get_attribute('href')) # getting the url of all watches
for url in URL_1: # loop for every watches in the list
    driver.get(url)
    
    try:
        rating = driver.find_element_by_xpath("//span[@data-hook='rating-out-of-text']") # collecting ratings
        Rating_1.append(rating.text)
        
    except NoSuchElementException as e:
        Rating_1.append("No data") # appending the No data if data is not is there
            
    try:
        return_ = driver.find_element_by_xpath("//div[@class='a-section a-spacing-none icon-content']") # collecting return
        Return_1.append(return_.text)
        
    except NoSuchElementException as e:
        Return_1.append("No data") # appending the No data if data is not is there
            
    try:
        avil = driver.find_element_by_xpath("//div[@id='availability']") # collecting avialability of the product
        Avia_1.append(avil.text)
        
    except NoSuchElementException as e:
        Avia_1.append("No data") # appending the No data if data is not is there
            
    try:
        details = driver.find_element_by_xpath("//div[@id='feature-bullets']") # collecting details of the product
        Details_1.append(details.text)
        
    except NoSuchElementException as e:
        Details_1.append("No data") # appending the No data if data is not is there
        
    try:
        no_ratings = driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']") # collecting number of ratings
        No_ratings_1.append(no_ratings.text)
    except NoSuchElementException as e:
        No_ratings_1.append("No data") # appending the No data if data is not is there
        
    try:
        price = driver.find_element_by_xpath("//span[@id='priceblock_ourprice']") # collecting price of the watch
        price_1.append(price.text)
    except NoSuchElementException as e:
        price_1.append("No data") # appending the No data if data is not is there
    
    try:
        delivery = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b[1]") # collecting expected delivery.
        Delivery_1.append(delivery.text)
    except NoSuchElementException as e:
        Delivery_1.append("No data") # appending the No data if data is not is there 
        
    try:
        brand = driver.find_element_by_xpath("//div[@class='a-section a-spacing-none']") # collect brand of watches.
        Brand_1.append(brand.text)
    except NoSuchElementException as e:
        Brand_1.append("No data") # appending the No data if data is not is there 
    
    try:
        product_name = driver.find_element_by_xpath("//span[@id='productTitle']") # collect product name.
        Product_name_1.append(product_name.text)
    except NoSuchElementException as e:
        Product_name_1.append("No data") # appending the No data if data is not is there 
        
print(URL_1)       
print(Rating_1)
print(Return_1)
print(Avia_1)
print(Details_1)
print(No_ratings_1)
print(price_1)
print(Delivery_1)  
print(Brand_1)
print(Product_name_1)
```

```python
driver.get("https://www.flipkart.com/") # open url

driver.maximize_window() # maximize driver window
time.sleep(4)  # sleep code for 4 second

#close login window
driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']").click()

# enter realme narzo 20 pro in search field
driver.find_element_by_name("q").send_keys("realme narzo 20 pro")

# click on search icon
driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()

driver.implicitly_wait(10) # wait for open all elements of the page
```

#### 7. Write a program to scrap all the available details of top 10 gaming laptops from digit.in.

```python
# open chromedriver using path
driver = webdriver.Chrome("I:\Software//chromedriver.exe")
```

```python
driver.get("https://digit.in/") # open url

driver.maximize_window() # maximize driver window
time.sleep(4)  # sleep code for 4 second

# click on laptops
driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[4]/ul/li[3]/a").click()
time.sleep(3)  # sleep code for 3 second

# click on best gaming laptops in india
driver.find_element_by_xpath("/html/body/div[6]/div/div[2]/div[2]/ul/li[10]/a").click()

driver.implicitly_wait(10) # wait for open all elements of the page
```

```python
graphics_processor=[]

graphics_pro = driver.find_elements_by_xpath("//div[@class='Spcs-details']/table[1]/tbody[1]/tr[9]/td[3]") # collecting graphics processor name
for i in graphics_pro:
    graphics_processor.append(i.text)
print(graphics_processor)
```

```python
URL=[]
model_name=[]
OS=[]
resolution_=[]
display_size=[]
ram_=[]
laptop_weight=[]
processor_model_name=[]
storage_=[]
price_=[]

urls = driver.find_elements_by_xpath("//div[@class='right-container']/div/a") # collect the urls of all the laptops
for i in urls:
    URL.append(i.get_attribute('href')) # getting the url of all laptops
for url in URL: # loop for every laptops in the list
    driver.get(url)
    
    price = driver.find_element_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[4]/div[1]/b") # collecting price
    price_.append(price.text)
        
    try:
        model = driver.find_element_by_xpath("//*[@id='specs-basic-information']/div[2]/table/tbody/tr[1]/td[3]") # collecting model names
        model_name.append(model.text)
        
    except NoSuchElementException as e:
        model_name.append("-") # appending the "-" if data is not there
        
    try:
        os = driver.find_element_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[3]/div/ul/li[1]/div") # collecting operating systems
        OS.append(os.text)
        
    except NoSuchElementException as e:
        OS.append("-") # appending the "-" if data is not there 
        
    try:
        driver.find_element_by_xpath("//div[@class='tab']/button[2]").click()
        resolution = driver.find_element_by_xpath("//*[@id='specs-display']/div[2]/table/tbody/tr[1]/td[3]") # collecting resolution
        resolution_.append(resolution.text)
        
    except NoSuchElementException as e:
        resolution_.append("-") # appending the "-" if data is not there
      
    try:
        dis_size = driver.find_element_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[3]/div/ul/li[2]/div") # collecting display size
        display_size.append(dis_size.text)
    except NoSuchElementException as e:
        display_size.append("-") #appending the "-" if data is not there
        
    try:
        driver.find_element_by_xpath("//div[@class='tab']/button[4]").click()
        ram = driver.find_element_by_xpath("//*[@id='specs-memory']/div[2]/table/tbody/tr[1]/td[3]") # collecting ram
        ram_.append(ram.text)
    except NoSuchElementException as e:
        ram_.append("-") #appending the "-" if data is not there
        
    try:
        driver.find_element_by_xpath("//div[@class='tab']/button[5]").click()
        weight = driver.find_element_by_xpath("//*[@id='specs-physical-specifications']/div[2]/table/tbody/tr[1]/td[3]") # collecting laptop weight
        laptop_weight.append(weight.text)
    except NoSuchElementException as e:
        laptop_weight.append("-") #appending the "-" if data is not there
        
    try:
        pro_name = driver.find_element_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[3]/div/ul/li[3]/div") # collecting processor model name
        processor_model_name.append(pro_name.text)
    except NoSuchElementException as e:
        processor_model_name.append("-") #appending the "-" if data is not there
        
    try:
        driver.find_element_by_xpath("//div[@class='tab']/button[7]").click()
        storage = driver.find_element_by_xpath("//*[@id='specs-storage']/div[2]/table/tbody/tr[2]/td[3]") # collecting storage capacity
        storage_.append(storage.text)
    except NoSuchElementException as e:
        storage_.append("-") #appending the "-" if data is not there

print(price_)
print(model_name)
print(OS)
print(resolution_)
print(display_size)
print(ram_)
print(laptop_weight)
print(processor_model_name)
print(storage_)
```

```python
# close driver
driver.close()
```

```python
df = pd.DataFrame({})
df['Model_name'] = model_name
df['Price'] = price_
df['Storage'] = storage_
df['Operating_system'] = OS
df['Resolution'] = resolution_
df['Display_size'] = display_size
df['Ram'] = ram_
df['Weight'] = laptop_weight
df['Processor_model_name'] = processor_model_name
df['Graphics_processor'] = graphics_processor
df
```
